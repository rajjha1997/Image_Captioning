# Initializing the COCO API to get the dataset

```python
import os
import sys
sys.path.append('/opt/cocoapi/PythonAPI')
from pycocotools.coco import COCO

# COCO API initialization for instance 
data_path = '/opt/cocoapi'
data_type = 'val2014'
annotatedFile_instance = os.path.join(data_path, 'annotations/instances_{}.json'.format(data_type))
coco = COCO(annotatedFile_instance)

# COCO API initialization for captions 
annotatedFile_captions = os.path.join(data_path, 'annotations/captions_{}.json'.format(data_type))
coco_caps = COCO(annotatedFile_captions)

# get image ids 
imaged_ids = list(coco.anns.keys())
```
